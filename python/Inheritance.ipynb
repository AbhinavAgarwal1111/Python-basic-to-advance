```python
#  Create a Python base class called Shape to represent basic shapes. 
#     The Shape class should have the following attributes and methods:

# Attributes:

# name (string): The name of the shape.
# Methods:

# __init__(self, name): A constructor that initializes the name attribute.
# area(self): A method that returns 0.0 (to be overridden by subclasses).
# Next, create two subclasses, Circle and Rectangle, that inherit from the Shape class. 
# These subclasses should override the area method to calculate the area of a circle and a rectangle, respectively.

# Circle Class:

# Attributes:
# radius (float): The radius of the circle.
# Methods:
# __init__(self, name, radius): A constructor that initializes the name and radius attributes.
# area(self): Override the area method to calculate and return the area of the circle using the formula: 
#     area = π * radius^2, where π (pi) is approximately 3.14159.
# Rectangle Class:

# Attributes:
# length (float): The length of the rectangle.
# width (float): The width of the rectangle.
# Methods:
# __init__(self, name, length, width): A constructor that initializes the name, length, and width attributes.
# area(self): Override the area method to calculate and return the area of the rectangle using the formula: area = length * width.

class Shape:
    def __init__(self, name):
        self.name = name
    
    def area(self):
        return 0.0

class Circle(Shape):
    def __init__(self,name,radius):
        super().__init__(name)
        self.radius = radius
    
    def area(self):
        areaa = 3.14159 * pow(self.radius,2)
        return areaa
    
class Rectangle(Shape):
    def __init__(self, name, length, width):
        super().__init__(name)
        self.name = name 
        self.length = length
        self.width = width
    
    def area(self):
        areaa = self.length * self.width
        return areaa

circle = Circle("Circle",5)
rectangle = Rectangle("Rectangle",5,6)

circle_area = circle.area()
rectangle_area = rectangle.area()

print(f"{circle.name} Area: {circle_area:.2f}")
print(f"{rectangle.name} Area: {rectangle_area:.2f}")
        
```

```python
# Create a Python base class called Animal to represent basic animal information. The Animal class should have the following attributes and methods:

# Attributes:

# name (string): The name of the animal.
# species (string): The species of the animal.
# Methods:

# __init__(self, name, species): A constructor that initializes the name and species attributes.
# make_sound(self): A method that returns "Some generic sound" (to be overridden by subclasses).
# Next, create two subclasses, Dog and Cat, that inherit from the Animal class. These subclasses should override the make_sound method to provide specific sounds for dogs and cats.

# Dog Class:

# Methods:
# __init__(self, name): A constructor that initializes the name attribute.
# make_sound(self): Override the make_sound method to return "Woof!"
# Cat Class:

# Methods:
# __init__(self, name): A constructor that initializes the name attribute.
# make_sound(self): Override the make_sound method to return "Meow!"
# Create instances of both the Dog and Cat classes, and then print the names, species, and sounds of the animals.

# Your task is to write the Python code for the Animal, Dog, and Cat classes, demonstrate inheritance, and print the information and sounds of the animals.

class Animal:
    def __init__(self, name, species):
        self.name = name
        self.species = species
    
    def make_sound(self):
        return "Some generic sound"
    
class Dog(Animal):
    def __init__(self, name):
        super().__init__(name, "Dog")
    
    def make_sound(self):
        return "Woof!"
    
class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Cat")
    
    def make_sound(self):
        return "Meow!"

dog = Dog("Kutta")
cat = Cat("Billi")

print(f"Animal Name: {dog.name}")
print(f"Animal Species: {dog.species}")
print(f"Animal Sound: {dog.make_sound()}")

print(f"\nAnimal Name: {cat.name}")
print(f"Animal Species: {cat.species}")
print(f"Animal Sound: {cat.make_sound()}")
```

```python
# Create a Python base class called Vehicle to represent basic vehicle information. The Vehicle class should have the following attributes and methods:

# Attributes:

# make (string): The make or manufacturer of the vehicle.
# model (string): The model of the vehicle.
# year (integer): The manufacturing year of the vehicle.
# Methods:

# __init__(self, make, model, year): A constructor that initializes the make, model, and year attributes.
# get_info(self): A method that returns a string containing information about the vehicle in the following format: "Make: [Make], Model: [Model], Year: [Year]".
# Next, create two subclasses, Car and Motorcycle, that inherit from the Vehicle class. These subclasses should add their own specific attributes and methods.

# Car Class:

# Attributes:
# num_doors (integer): The number of doors in the car.
# Methods:
# __init__(self, make, model, year, num_doors): A constructor that initializes the make, model, year, and num_doors attributes.
# get_info(self): Override the get_info method to include the number of doors in the information string.
# Motorcycle Class:

# Attributes:
# engine_size (string): The engine size of the motorcycle.
# Methods:
# __init__(self, make, model, year, engine_size): A constructor that initializes the make, model, year, and engine_size attributes.
# get_info(self): Override the get_info method to include the engine size in the information string.
# Create instances of both the Car and Motorcycle classes, and then print the information about the vehicles, including their make, model, year, and any additional attributes.

# Your task is to write the Python code for the Vehicle, Car, and Motorcycle classes, demonstrate inheritance, and print the information about the vehicles.

class Vehical:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
    
    def get_info(self):
        return "Make: f{self.make}, Model: {self.model}, Year: {self.year}"

class Car(Vehical):
    def __init__(self, make, model, year, num_doors):
        super().__init__( make, model, year)
        self.num_doors = num_doors
        
    def get_info(self):
        return f"Make: {self.make}, Model: {self.model}, Year: {self.year}, Number of Doors: {self.num_doors}"
    
class Motorcycle(Vehical):
    def __init__(self, make, model, year, engine_size):
        self.engine_size = engine_size
        super().__init__( make, model, year)
        
    def get_info(self):
        return f"Make: {self.make}, Model: {self.model}, Year: {self.year}, Number of Doors: {self.engine_size}"
    
car = Car("Toyota", "Camry", 2022, 4)
motorcycle = Motorcycle("Honda", "CBR500R", 2023, "500cc")

print("Car Information:")
print(car.get_info())

print("\nMotorcycle Information:")
print(motorcycle.get_info())
```
