```python
# Create a Python class called Rectangle that represents a rectangle. The class should have the following attributes and methods:

# Attributes:

# width (integer): The width of the rectangle.
# height (integer): The height of the rectangle.
# Methods:

# __init__(self, width, height): A constructor that initializes the width and height attributes.
# area(self): A method that calculates and returns the area of the rectangle (width * height).
# perimeter(self): A method that calculates and returns the perimeter of the rectangle (2 * width + 2 * height).
# After defining the Rectangle class, create an instance of it and use its methods to calculate and print the area and perimeter 
# of a rectangle with a width of 5 units and a height of 3 units.

# Your task is to write the Python code for the Rectangle class and demonstrate its usage to calculate the area and perimeter.

class rectangle:
    def __init__(self,width,height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

Rectangle = rectangle(5,3)
print("Area of rectangle is :",Rectangle.area())
print("Perimeter of rectangle is :",Rectangle.perimeter())
```

```python
# Create a Python class called Person to represent a person's information. The class should have the following attributes and
# methods:

# Attributes:

# name (string): The person's name.
# age (integer): The person's age.
# Methods:

# __init__(self, name, age): A constructor that initializes the name and age attributes.
# greet(self): A method that prints a greeting message, including the person's name and age. For example,
#     "Hello, my name is [Name], and I am [Age] years old."
# After defining the Person class, create an instance of it and use its methods to print a greeting message for 
# a person with the name "Alice" and age 25.

# Your task is to write the Python code for the Person class and demonstrate its usage to print a greeting message.

class Person:
    def __init__(self,name,age):
        self.name = str(name)
        self.age = int(age)
        
    def greet(self):
        print(f"Hello, my name is {self.name}, and I am {self.age} years old.")
        
greeting = Person("Alise",25)

greeting.greet()
```

```python
# Create a Python class called BankAccount to represent a simple bank account. The class should have the following attributes
# and methods:

# Attributes:

# account_number (string): A unique account number for the bank account.
# balance (float): The current balance in the account.
# Methods:

# __init__(self, account_number): A constructor that initializes the account_number attribute.
#     The initial balance should be set to 0.0.
# deposit(self, amount): A method that allows a user to deposit a specified amount into the account.
#     It should update the balance.
# withdraw(self, amount): A method that allows a user to withdraw a specified amount from the account if the balance is 
#    sufficient. It should update the balance.
# get_balance(self): A method that returns the current balance of the account.
# After defining the BankAccount class, create an instance of it, perform a series of deposits and withdrawals, 
# and print the final balance of the account.

# Your task is to write the Python code for the BankAccount class and demonstrate its usage to perform transactions 
# and check the balance.

class BankAccount:
    def __init__(self, account_number):
        self.account_number = account_number
        self.balance = 0.0
        
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f} into the account {self.account_number}")
            
    def withdraw(self, amount):
         if amount > 0 and amount <= self.balance:
                self.balance -= amount
                print(f"Withdrawn ${amount:.2f} from account {self.account_number}")
            
    def get_balance(self):
        return self.balance
    
account = BankAccount("9891310")

account.deposit(1000.50)
account.withdraw(500.25)
account.deposit(200.75)
account.withdraw(1500.00)

final_balance = account.get_balance()
print(f"Final balance in account {account.account_number}: ${final_balance:.2f}")
```

```python
# Create a Python class called Student to represent student information. The class should have the following attributes 
# and methods:

# Attributes:

# name (string): The student's name.
# roll_number (string): The student's roll number.
# marks (dictionary): A dictionary where keys are subject names (strings) and values are the corresponding marks (integers) 
#     obtained by the student in those subjects.
# Methods:

# __init__(self, name, roll_number): A constructor that initializes the name, roll_number, and an empty marks dictionary for
#     the student.
# add_subject_mark(self, subject, mark): A method that allows adding a subject and its mark to the student's marks dictionary.
# get_total_marks(self): A method that calculates and returns the total marks obtained by the student by summing up all the marks
#     in the marks dictionary.
# get_average_marks(self): A method that calculates and returns the average marks obtained by the student by dividing the total
#     marks by the number of subjects.
# After defining the Student class, create an instance of it, add marks for three subjects, and then print the student's 
# name, roll number, total marks, and average marks.

# Your task is to write the Python code for the Student class and demonstrate its usage to manage and calculate student marks.

class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        self.marks = {}
    
    def add_subject_mark(self,subject,mark):
        self.marks[subject] = mark
    
    def get_total_marks(self):
        return sum(self.marks.values())
    
    def get_average_marks(self):
        total_marks = self.get_total_marks()
        num_subject = len(self.marks)
        if num_subject == 0:
            return 0.0
        return total_marks/num_subject

student = Student("Alice", "12345")

student.add_subject_mark("Math", 95)
student.add_subject_mark("Science", 88)
student.add_subject_mark("History", 92)

total_marks = student.get_total_marks()
average_marks = student.get_average_marks()

print(f"Student Name: {student.name}")
print(f"Roll Number: {student.roll_number}")
print(f"Total Marks: {total_marks}")
print(f"Average Marks: {average_marks:.2f}")
        
```

```python
# Create a Python class called Employee to represent employee information.
# The class should have the following attributes and methods:

# Attributes:

# employee_id (string): The unique employee ID.
# name (string): The name of the employee.
# salary (float): The monthly salary of the employee.
# Methods:

# __init__(self, employee_id, name, salary): A constructor that initializes the employee_id, name, and salary attributes.
# calculate_yearly_salary(self): A method that calculates and returns the yearly salary of the employee 
#     (12 times the monthly salary).
# After defining the Employee class, create an instance of it, and then print the employee's ID, name, monthly salary, and 
# yearly salary.

# Your task is to write the Python code for the Employee class and demonstrate its usage to calculate and display employee
# information.

class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_salary(self):
        yearly_salary = 12 * self.salary
        return yearly_salary
    
emp = Employee("1111","Abhinav Agarwal",1800000)

employee_id = emp.employee_id
name = emp.name
monthly_salary = emp.salary
salary_pa = emp.calculate_yearly_salary()

print(f"Employee id is {employee_id}")
print(f"Employee Name is {name}")
print(f"Employee Salary is ${monthly_salary:.2f}")
print(f"Employee Yearly Salary is ${salary_pa:.2f}")

```

```python
# Create a Python class called Car to represent car information. The class should have the following attributes and methods:

# Attributes:

# make (string): The make or manufacturer of the car.
# model (string): The model of the car.
# year (integer): The manufacturing year of the car.
# mileage (float): The current mileage of the car in miles.
# Methods:

# __init__(self, make, model, year, mileage): A constructor that initializes the make, model, year, and mileage attributes.
# update_mileage(self, new_mileage): A method that allows updating the current mileage of the car to a new value.
# get_car_info(self): A method that returns a string containing information about the car in the following format:
#         "Make: [Make], Model: [Model], Year: [Year], Mileage: [Mileage] miles."
# After defining the Car class, create an instance of it, update the mileage, and then print the car's
# make, model, year, and mileage.

# Your task is to write the Python code for the Car class and demonstrate its usage to manage and display car information.

class car:
    def __init__(self, make, model, year, mileage):
        self.make = make 
        self.model = model
        self.year = year
        self.mileage = mileage
    
    def update_mileage(self, new_mileage):
        if new_mileage >= 0:
            self.mileage = new_mileage
        
    def get_car_info(self):
        print(f"Make: {self.make}, Model: {self.model}, Year: {self.year}, Mileage: {self.mileage:.2f} miles.")
    
car_info = car("audi","Q7","2019",63)    
car_info.update_mileage(25000)
print(car_info.get_car_info())
```

```python
# Create a Python class called Circle to represent a circle. The class should have the following attributes and methods:

# Attributes:

# radius (float): The radius of the circle.
# Methods:

# __init__(self, radius): A constructor that initializes the radius attribute.
# calculate_area(self): A method that calculates and returns the area of the circle using the formula:
#         area = π * radius^2, where π (pi) is approximately 3.14159.
# calculate_circumference(self): A method that calculates and returns the circumference of the circle using the formula:
#         circumference = 2 * π * radius.
# After defining the Circle class, create an instance of it with a radius of 5.0 units, and then calculate and print
# the area and circumference of the circle.

# Your task is to write the Python code for the Circle class and demonstrate its usage to calculate and 
# display circle information.

class Circle:
    def __init__(self, radius):
        self.radius = radius
        
    def calculate_area(self):
        return 3.14159 * pow(self.radius,2)
    
    def calculate_circumference(self):
        return  2 * 3.14159 * self.radius
    
circle = Circle(6)
area = circle.calculate_area()
circumference = circle.calculate_circumference()

print(f"Area: of Circle is {area:.2f} & Circumference: of the circle is {circumference:.2f}")
```
